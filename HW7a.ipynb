```python
import tweepy
import numpy as np
import pandas as pd
import json
import numpy as np
from datetime import datetime
import matplotlib.pyplot as plt

# Import and Initialize Sentiment Analyzer
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()

# Twitter API Keys
from config import (consumer_key, 
                    consumer_secret, 
                    access_token, 
                    access_token_secret)

# Setup Tweepy API Authentication
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)
api = tweepy.API(auth, parser=tweepy.parsers.JSONParser())
```

```python


target_terms = ("@nytimes", "@cnnbrk","@cbsnews","@foxnews","@bbcbreaking")




results_list = []
results_list2 = []



for target in target_terms:

    compound_list = []
    positive_list = []
    negative_list = []
    neutral_list = []
    oldest_tweet = None
    counter = 0

    public_tweets = api.user_timeline(target,count=100, max_id = oldest_tweet)
    for tweet in public_tweets:
        oldest_tweet = tweet["id"] - 1
        counter = counter + 1
        print (tweet["text"])
        results = analyzer.polarity_scores(tweet["text"])
        compound = results["compound"]
#         pos = results["pos"]
#         neu = results["neu"]
#         neg = results["neg"]
        
        sentiment = {
        "User": target,
        "Compound": (compound),
#         "Positive": (pos),
#         "Neutral": (neg),
#         "Negative": (neu),
        "tweetago": counter
          }
        results_list.append(sentiment)



#         Add each value to the appropriate list
        compound_list.append(compound)
#         positive_list.append(pos)
#         negative_list.append(neg)
#         neutral_list.append(neu)

    sentimentmean = {
    "User": target,
    "Compound": np.mean(compound_list),
#     "Positive": np.mean(positive_list),
#     "Neutral": np.mean(negative_list),
#     "Negative": np.mean(neutral_list),
    "Tweet Count": len(compound_list)
        
      }
    results_list2.append(sentimentmean)
    
```

```python
news_df = pd.DataFrame(results_list).set_index("User").round(3)

news_df.to_csv("outputdata/newssentiment.csv")
news_df.head()
```

```python
newsmean_df = pd.DataFrame(results_list2).set_index("User").round(3)

newsmean_df.to_csv("outputdata/newsmeansentiment.csv")
newsmean_df
```

```python

news_dfi=news_df.reset_index()
nwm=news_dfi.pivot(index='tweetago',columns='User')['Compound']

nwm=nwm.rename(columns={'@cbsnews':'CBS','@nytimes':'NYT','@cnnbrk':"CNN","@bbcbreaking":"BBC","@foxnews":"FOX"})
nwm.head()
```

```python
x=nwm.index
cp=nwm['CBS']

ip=nwm['NYT']
cn=nwm['CNN']
bc=nwm['BBC']
fx=nwm['FOX']



fig1, ax = plt.subplots()

ax.scatter(x, cp, marker="o",color = 'red', alpha = 0.75)

ax.scatter(x, ip,marker="o",color = 'blue', alpha =0.75)
ax.scatter(x, cn, marker="o",color = 'Orange', alpha = 0.75)

ax.scatter(x, bc,marker="o",color = 'green', alpha =0.75)
ax.scatter(x, fx, marker="o",color = 'yellow', alpha = 0.75)


ax.legend(loc=10, bbox_to_anchor=(1.18, .5), ncol=1, title="Media Outlets")

# ax.legend(loc="best")

now = datetime.now()
now = now.strftime("%Y-%m-%d %H:%M")
plt.title(f"Sentiment Analysis of Tweets ({now})")

plt.ylabel("Tweet Polarity")
plt.xlabel("Tweets Ago")
plt.tight_layout()
plt.grid()
plt.savefig("images/mediasentiments.png")
plt.show()


```
